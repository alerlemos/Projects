```python
from selenium import webdriver
import pandas as pd
import numpy as np
from selenium.webdriver.common.keys import Keys
import requests
import bs4
import lxml
```

```python
# Preciso conseguir o url correto
url = 'https://p-nt-www-amazon-com-br-kalias.amazon.com.br/1984-George-Orwell-ebook/dp/B009XE662U/ref=sr_1_1?__mk_pt_BR=%C3%85M%C3%85%C5%BD%C3%95%C3%91&dchild=1&keywords=B009XE662U&qid=1622505480&s=digital-text&sr=1-1'
result = requests.get(url)
```

```python
soup = bs4.BeautifulSoup(result.text,"lxml")
print(soup)
```

```python
x = soup.select("div")
```

```python
X = []
for xx in x: 
    try:
        #print(xx['id'])
        if xx['id'] == 'detailBullets_feature_div':
            X.append(xx)
    except:
        pass
```

```python
i = 0
for l in X[0].select('li'):
    print(l.text.replace('\n',''))
    #print(i)
    i += 1
```
